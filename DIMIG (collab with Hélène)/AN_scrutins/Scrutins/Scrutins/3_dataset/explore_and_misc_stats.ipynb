### 1. Data

````python
import pandas as pd

# Load with low_memory=False to avoid mixed type warnings
df = pd.read_csv('scrutins_complete.csv', low_memory=False)
print(df.shape)
df.head()
````

### Pour alléger et parce que l'étude porte sur la période 2014-2024, on enlève tous les votes antérieurs à 2014

````python
# Filter: keep only votes from 2014 to 2024
print(df.shape[0])
old = df.shape[0]
df = df[(df['scrutin_date'] >= '2014-01-01') & ('2024-12-31' >= df['scrutin_date'])]
print(f"After filtering (>= 2014): {df.shape[0]:,} rows. We drop {(old - df.shape[0]):,} rows")
df.sort_values('scrutin_date', ascending=True)["scrutin_date"].head()
````
